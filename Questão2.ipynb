```
# Importação e tratamento dos dados
planta = open('/content/drive/MyDrive/Colab Notebooks/respostaSMA.csv').readlines()

num = np.array([279.31])
dem = np.array([3.61,1])
G = ctl.tf(num, dem)

#Plotando o gráfico
x,y = ctl.step(0.29*G)
plt.plot(y,x, label='Modelo matemático')

t = []
w = []
for i in range(len(planta)):
  if i>0: #ignora linhas 0 
    linha = planta[i].split(",")
    t.append(float(linha[0]))
    w.append(float(linha[1]))
t = np.array(t)
w = np.array(w)

#Plotando o gráfico
plt.plot(t,w-3062, label='Modelo experimental')
plt.xlabel('t, s')
plt.ylabel('$w$, rpm')
plt.legend(loc='best')
plt.grid()
plt.xlim(0,30)
plt.title('Modelo matemático e Modelo experimental')
plt.show()
```
